```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years = df1.columns[1:]
total_values=df1[years].sum()
plt.figure(figsize=(30, 30))
plt.barh(years,total_values,edgecolor="white")
plt.xlabel("Total values", size= 20)
plt.ylabel("Years", size= 20)
plt.title("Total Values Per Year", size =30)
plt.show()
```

```python
top_20_countries = df1.nlargest(20, '1960')

plt.figure(figsize=(12, 10))
plt.bar(top_20_countries['Country Name'], top_20_countries['1960'],color="green")
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 20 Most Populated Countries in 1960')
plt.xticks(rotation=60)
plt.show()
```

```python
top_20_countries = df1.nlargest(20, '2023')

plt.figure(figsize=(12, 10))
plt.bar(top_20_countries['Country Name'], top_20_countries['2023'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 20 Most Populated Countries in 2023')
plt.xticks(rotation=60)
plt.show()
```

## Metadata Country

```python
import pandas as pd


data = pd.read_csv(r"Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_1045708.csv")
data
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python

region_counts = data['Region'].value_counts()


plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar', color='orange')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.title('Number of Countries by Region')
plt.xticks(rotation=50)
plt.tight_layout()
plt.show()
```

```python
income_groups = data.groupby('IncomeGroup').size().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
income_groups.plot(kind='bar', color='red')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.title('Number of Countries in Each Income Group')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
